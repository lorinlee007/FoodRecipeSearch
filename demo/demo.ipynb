# Evaluation BM25 vs FAISS: Keyword Search Queries

```python
import sys
sys.path.append("../src")

import search_bm25
import search_faiss

bm25_searcher = search_bm25.get_searcher()
faiss_searcher, faiss_meta = search_faiss.get_searcher()
```

```python
queries = ["baked salmon with lemon", "texas sheet cake", "low fat chicken", "molasses ginger cookies", "broccoli cheddar soup"]

for query in queries:
    results = search_bm25.search(query, searcher=bm25_searcher)

    for r in results:
        print(f"{r['rank']:2d}. {r['score']:.3f} | {r['id']} | {r['name']}")
    print()
```

```python
for query in queries:
    results = search_faiss.search(query, searcher=faiss_searcher, metadata=faiss_meta)
    rank = 1
    for r in results:
        print(f"{rank:2d}. {r['score']:.3f} | {r['id']} | {r['name']}")
        rank += 1
    print()
```
